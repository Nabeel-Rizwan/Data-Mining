```python
from google.colab import files

uploaded=files.upload()
data = pd.read_excel('Naive Bayes.xlsx')
```

```python
# Assuming your data has columns for features and a column for the target variable
# Separate features and target variable

X = data[['color', 'type', 'origin']]  # Replace with your feature column names
y = data['stolen']  # Replace with your target column name

# Calculate class probabilities (P(class))

class_probabilities = y.value_counts(normalize=True)  # Calculates the probabilities of each class occurring in the target variable y

# Calculate conditional probabilities (P(feature | class))

conditional_probs = {}  # Initializes an empty dictionary to store conditional probabilities.

for feature in X.columns:
    for cls in class_probabilities.index:   # Iterates through each feature and class.
        mask = y == cls                     # Creates a mask to filter rows based on the current class
        conditional_probs[(feature, cls)] = X.loc[mask, feature].value_counts(normalize=True)   # Calculates the conditional probabilities of each feature given the class.

# Implementing Naive Bayes Classifier (for example, using simple counting)

def naive_bayes_classification(sample):
    probabilities = {}
    for cls in class_probabilities.index:
        probabilities[cls] = class_probabilities[cls]
        for feature in sample.index:
            probabilities[cls] *= conditional_probs.get((feature, cls), {}).get(sample[feature], 0)
    return max(probabilities, key=probabilities.get)



new_sample = pd.Series({'color': 'red', 'type': 'suv', 'origin': 'domestic'})  # Replace values

predicted_class = naive_bayes_classification(new_sample)

print(f"Predicted Class: {predicted_class}")
```
