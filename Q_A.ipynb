```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Define the questions and answers

questions = [
    "What are the courses in JMI ",
    "What is JMI ",
    "how is btech in JMI ",
    "Tell me about computer engineering department",
    "How are the placements in JMI ",
    "Tell the fullform of JMI ",
    "where is JMI located",
]

answers = [
    "The courses in JMI are Btech, Mass Communication, BArch, BCom, Etc. It is a central universiy. Therefore, Almost all courses are available. ",
    "JMI is a central university situated in Delhi",
    "BTech is very good in JMI. It is known for the best engineering. In BTech, it offers COE, CE, ME, ECE,EE. ",
    "computer engineering department is the best in JMI. It has best placements among BTech ",
    "Placements are awesome",
    "JMI Fullform is Jamia Millia Islamia ",
    "JMI is located in new delhi",
]

vectorizer = CountVectorizer()

questions_defined = vectorizer.fit_transform(questions)

user_question = input(" What is your question? Make sure your question is related to JMI Domain \n")

user_question_vec = vectorizer.transform([user_question])

similarities = cosine_similarity(user_question_vec, questions_defined)

most_similar_index = np.argmax(similarities)

most_similar_answer = answers[most_similar_index]

print("User Question:", user_question)

print("Most Similar Question:", questions[most_similar_index])

print("Generated Answer:", most_similar_answer)
```
