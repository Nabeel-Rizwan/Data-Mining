```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded=files.upload()
```

```python
data = pd.read_excel('txt2.xlsx')
# data.head()
print(data)
```

```python
datasubset = data.loc[:, ["Age","Annual Income"]]   # : selects all rows in the attributes
plt.figure(figsize=(10, 7))   # represents the height and width of the graph, width and height
plt.scatter(datasubset[["Age"]], datasubset["Annual Income"], s=100, c='blue')    # marker size s = 100
plt.show()
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(10, 7))

dendrogram = sch.dendrogram(sch.linkage(datasubset, method  = "ward"))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Select columns for clustering (adjust column names as needed)
# For example, use columns 'X' and 'Y' as features for clustering

X = data[['Age', 'Annual Income']].values  # Replace with your column names

# Number of clusters
n_clusters = 3  # Change this according to your requirement

# Perform Agglomerative Clustering
model = AgglomerativeClustering(n_clusters=n_clusters)
labels = model.fit_predict(X)

# Print cluster labels for each data point
print("Cluster Labels:")
print(labels)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.title('Agglomerative Clustering')
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.show()
```
