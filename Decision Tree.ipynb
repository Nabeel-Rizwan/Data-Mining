```python
def calculate_gini_index_for_column(dataset, column_name):
    total_samples = len(dataset)
    attribute_values = dataset[column_name]
    class_labels = dataset.iloc[:, -1]  # Assuming the last column is the target column
    unique_values = attribute_values.unique()
    gini_index_sum = 0

    for value in unique_values:
        filtered_data = dataset[attribute_values == value]
        label_counts = filtered_data.iloc[:, -1].value_counts()
        label_probabilities = label_counts / len(filtered_data)
        gini = 1 - sum(label_probabilities ** 2)
        gini_index_sum += (len(filtered_data) / total_samples) * gini

    return gini_index_sum

# Calculate Gini index for all columns (attributes)

gini_indices = {}
for column in data.columns[:-1]:  # Exclude the last column (target class)
    gini_index = calculate_gini_index_for_column(data, column)
    gini_indices[column] = gini_index

print("Gini Index for each attribute:")
for attribute, gini in gini_indices.items():
    print(f"{attribute}: {gini}")
```
