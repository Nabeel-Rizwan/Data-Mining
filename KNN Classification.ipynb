```python
import math

K = 5
df['Distance'] = ""

# Calculate the number of rows in the DataFrame
num_rows = len(df)

for i in range(num_rows):
    # Assuming 'x' and 'y' are column names in your DataFrame
    distance = math.sqrt((df.loc[i, 'x'] - X_test) ** 2 + (df.loc[i, 'y'] - Y_test) ** 2)
    df.loc[i, 'Distance'] = distance
```

```python
df=df.sort_values(by=['Distance'])

data_new = pd.DataFrame()
data_new=df.iloc[0:K,:]
data_new

plt.scatter(df['x'],df['y'],c=df['label'])
plt.scatter(X_test,Y_test,c='r')
plt.scatter(data_new['x'],data_new['y'],c='g')
plt.show()
```
